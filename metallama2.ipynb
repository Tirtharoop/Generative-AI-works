```python
model_name_or_path = "TheBloke/Llama-2-13B-chat-GGML"
model_basename = "llama-2-13b-chat.ggmlv3.q5_1.bin"  # the model is in bin format
```

```python
from huggingface_hub import hf_hub_download
```

```python
from llama_cpp import Llama
```

```python

model_path = hf_hub_download(repo_id=model_name_or_path, filename=model_basename)
```

```python
lcpp_llm = None
lcpp_llm = Llama(
    model_path=model_path,
    n_threads=2, # CPU cores
    n_batch=512, # Should be between 1 and n_ctx, consider the amount of VRAM in your GPU.
    n_gpu_layers=32 # Change this value based on your model and your GPU VRAM pool.
    )
```

```python
# See the number of layers in GPU
lcpp_llm.params.n_gpu_layers
```

```python
#prompt = "give me a interface for your code where user will give input in HTML"
prompt = input("KI CHAI BOL: ")
prompt_template=f'''SYSTEM: You are a helpful, respectful and honest assistant. Always answer as helpfully.

USER: {prompt}

ASSISTANT:
'''

response=lcpp_llm(prompt=prompt_template, max_tokens=256, temperature=0.5, top_p=0.95,
                  repeat_penalty=1.2, top_k=150,
                  echo=True)

print(response["choices"][0]["text"])
```

```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.prompts.prompts import SimpleInputPrompt
```

```python
documents=SimpleDirectoryReader("/content/NLP").load_data()
documents
```

```python
# system_prompt="""
# You are a Q&A assistant. Your goal is to answer questions as
# accurately as possible based on the instructions and context provided.
# """

system_prompt="""
You are a code analist expert. Your goal is to provide a proper documentation of a given code to make it easily understandable.
Documentaion may include code analysis, complexity analysis, remarks.
"""

## Default format supportable by LLama2
query_wrapper_prompt=SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
!huggingface-cli login
```

```python
import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=1024,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16 , "load_in_8bit":True}
)
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.core import ServiceContext
from llama_index.embeddings.langchain import LangchainEmbedding

embed_model=LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))
```

```python
service_context=ServiceContext.from_defaults(
    chunk_size=5024,
    llm=llm,
    embed_model=embed_model
)
```

```python
index=VectorStoreIndex.from_documents(documents,service_context=service_context)

print(index)
```

```python
query_engine=index.as_query_engine()
```

```python
response=query_engine.query("Generate documentation based on the given code.")

print(response)
```
